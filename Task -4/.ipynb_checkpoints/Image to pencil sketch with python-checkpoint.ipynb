### Task 4- Image to Pencil Sketch with Python

This is a short project aimed at converting RGB images into pencil sketches. The main library that we are going to use for this purpose is OpenCV.

### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Loading and Plotting the Original Image

Now that the libraries have been imported, we will be loading our target image that needs to be converted to a pencil sketch. For this we will be using imread function of cv2 library which takes the path of the file as the sole argument.

We will also be converting the image form its original format to RGB format for better processing.

```python
img = cv2.imread('image2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

The above plot shows the image that we have imported and will be using for further processing.

### Converting the Original Image to Grayscale

To decrease the pixel complexity of image and increase the handling ease we will be converting this image to a grayscale format using cvtColor function.

```python
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_grayscale, cmap='gray')
plt.axis('off')
plt.title('GrayScale Image')
plt.show()
```

We can clearly see the grayscale image plotted above. 

### Inversion of Image

Next step in processing the image is to invert the grayscale so that we can study the granularity and pixel spread of the image with much more precision and detail.

We are doing this because in order to convert an image to a pencil sketch, we need to study the image with more accuracy and detail. Only then the sketch generation will be accurate and good enough to show any semblance of the original image.

```python
img_invert = cv2.bitwise_not(img_grayscale)
plt.figure(figsize=(8,8))
plt.imshow(img_invert, cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

Above, we can see the inverted image.

### Making inverted image smoother

We will be smoothing the inverted image to make sure that the sketch that we obtain in the end is less edgy and more smooth.

```python
img_smooth = cv2.GaussianBlur(img_invert, (21,21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smooth, cmap='gray')
plt.axis('off')
plt.title('Smooth Image')
plt.show()
```

The smoother version of the inverted image is obtained above.

### Converting the Smooth Image to Pencil Sketch

We are finished we image processing and the final step that's left is to covert it to a pencil sketch.

We know that cv2 library stores the image in terms of pixels in the form of arrays. Since array contains numbers for all pixels, we can manipulate any image imported using cv2 easily with the help of a simple python code.

For converting the smoother version of inverted image to a pencil sketch, we will be dividing the grayscale image array with smooth image array and scaling the output.

```python
pencil_sketch = cv2.divide(img_grayscale, 255-img_smooth, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')
plt.title('Final Pencil Sketch Image')
plt.show()
```

### Displaying all Images

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_grayscale,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smooth,cmap="gray")
plt.axis("off")
plt.title("Smooth Image")
plt.subplot(1,5,5)
plt.imshow(pencil_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```

In the result above, we can clearly see the conversion flow of the original image to a pencil sketch.

Hence the image has been successfully converted to a pencil sketch and the python code if reliably working.
